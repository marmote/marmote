## Reference pencil lead break

First let's read in the reference pencil lead break data for BreakTest_AluminiumBeam measurement.

```python
from helperfuncs.Raw_Data.ReadAEMeas import ReadAEMeas

file = '../../../Measurements/BreakTest_AluminiumBeam/penbreak-test.bin'
y, y2, T, fnames, start_time = ReadAEMeas(file)

Fs = 1/T
```

Calculate TDOA with simple threshold crossing method.

```python
from helperfuncs.Onset_Time.CalculateTDOA import CalculateTDOA
from helperfuncs.Onset_Time.Onset_Threshold import Onset_Threshold

AE_start, AE_start2, TD_meas, dummy, dummy = CalculateTDOA(y, y2, T, Onset_Threshold)
```

Plot AE data in separate axes.

```python
from helperfuncs.Onset_Time.PlotAEMeas import PlotAEMeas

PlotAEMeas(y,   y2,   AE_start*1e3, AE_start2*1e3,   
           T*1e3,  
           x_lim_min=0, x_lim_max=0.003*1e3, y_lim_min=-1, y_lim_max=1, 
           xlabel='time [msec]', ylabel='Sample []')
```

```python
from helperfuncs.Raw_Data.FunctionDisplayRecording_total import FunctionDisplayRecording_total

FunctionDisplayRecording_total(file)
```
