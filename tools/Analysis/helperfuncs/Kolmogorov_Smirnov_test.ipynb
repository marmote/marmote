# Kolmogorov–Smirnov test of uniformity of the numpy pseudo random number generator

## Introduction

The goal of this notepad is twofold. Firstly, it tries to answer the question how reliable the uniform pseudo random number generator is that comes with numpy. Secondly, it shows an example of using the built-in Kolmogorov-Smirnov test. 

The approach taken here is the following: Using the pseudo random number generator sample points are generated on the z=0 plane in a circle of unit radius. The Cartesian coordinates are subsequently transformed to Polar coordinates. These coordinates are independent and their Cumulative Distribution Functions are well-known. At this point the Kolmogorov-Smirnov test is employed to verify the null hypothesis, namely, that the original samples were indeed generated by a uniform random number generator. This process is repeated several times for different sample set sizes. 

## Random samples on a unit radius disc

First, uniform distribution random samples are generated on a unit radius disc. The x and y coordinates are independently generated with a [0, 1) range uniform generator. Samples not in the unit circle are rejected. 

```python
import numpy as np

def Generate_Disc( N, R ):
    x_list = np.array([])
    y_list = np.array([])

    while len(x_list) < N:
        x = (2*np.random.rand()-1)*R
        y = (2*np.random.rand()-1)*R
        if np.sqrt(x**2 + y**2) > R:
            continue
   
        x_list = np.append(x_list, x)
        y_list = np.append(y_list, y)

    return x_list, y_list
```

```python
R = 1;
N = 1e5;

x, y = Generate_Disc(N, R)
```

With x, y generated the polar coordinates r, alpha are calculated. These are also independent random variables, but while alpha's CDF is linear, the r's CDF is squared.

```python
from Statistics.Plot_2D_Hist_In_3D import Plot_2D_Hist_In_3D

r = np.sqrt(x**2 + y**2)
alpha = np.arctan2(y,x)

#print r
#print alpha

Plot_2D_Hist_In_3D(alpha, r, linspace(-4.5,4.5,1e2), linspace(-0.3,1.3,1e2),
                  'Histogram (normalized), N = %d'%(len(x)), 'alpha', 'r', 'Probability density')
```

## Kolmogorov-Smirnov test

In the following Kolmogorov-Smirnov test is applied on both r and alpha independently. The test returns the Type I error: The probability of being wrong if the uniformity hypothesis is rejected.

```python
from Statistics.Get_Empirical_CDF import Get_Empirical_CDF

# Angle - alpha
x, emp_CDF = Get_Empirical_CDF(alpha)
alpha_CDF_func = lambda x : x/(2*np.pi) + 0.5
CDF = alpha_CDF_func(x)

fig, axarr = plt.subplots(1, 1, sharex=True)
axarr.plot(x, emp_CDF, '.-b', x, CDF, '.-r')
axarr.set_title('CDFs of angles')
axarr.set_xlabel('angle')
axarr.set_ylabel('probability')
axarr.legend( ('Empirical CDF', 'CDF'), loc=4 )
```

```python
from scipy.stats import kstest

D, p = kstest(alpha, alpha_CDF_func)
#print D
#print p

print 'When rejecting uniformity, based on angles, the probability of being wrong (Type I error): %.2f%%\n'%p
```

```python
from Statistics.Get_Empirical_CDF import Get_Empirical_CDF

# Radius - r
x, emp_CDF = Get_Empirical_CDF(r)
r_CDF_func = lambda x : x**2
CDF = r_CDF_func(x)

fig, axarr = plt.subplots(1, 1, sharex=True)
axarr.plot(x, emp_CDF, '.-b', x, CDF, '.-r')
axarr.set_title('CDFs of radii')
axarr.set_xlabel('radius')
axarr.set_ylabel('probability')
axarr.legend( ('Empirical CDF', 'CDF'), loc=4 )
```

```python
from scipy.stats import kstest

D, p = kstest(r, r_CDF_func)
#print D
#print p

print 'When rejecting uniformity, based on radii, the probability of being wrong (Type I error): %.2f%%\n'%p
```

Let's look at uniformity as a function of sample size.

```python
err_alpha = []
err_r = []
Ns = []

for ii in xrange(1,6):
    N = 10**ii
    Ns.append(N)
    
    x, y = Generate_Disc(N, R)
    
    r = np.sqrt(x**2 + y**2)
    alpha = np.arctan2(y,x)
    
    D, p = kstest(alpha, alpha_CDF_func)
    err_alpha.append(p)
    
    D, p = kstest(r, r_CDF_func)
    err_r.append(p)    

fig, axarr = plt.subplots(1, 1, sharex=True)
axarr.semilogx(Ns, err_alpha, '.-b', Ns, err_r, '.-r')
axarr.set_title('Probability of being wrong when rejecting uniformity using Kolmogorov-Smirnov test')
axarr.set_xlabel('Sample size')
axarr.set_ylabel('probability')
axarr.legend( ('Angle','Radii'), loc=4 )
```

## Conclusion

The results show that using the above described approach to verify uniformity the sample sets can not always safely be considered originating form a valid uniform source. Also, increasing the sample set size does not guarantee better (more uniform) results.
