# Closer look on Variances

.

```python
N = 1000;
T = 1/750e3;
```

```python
starttime = 0.2
ConstMultStart = 2
ConstMultStop = 11

mu, sigma = 0, 1

#################################################
ym = mu + sigma*np.random.randn(N)

startidx = int(np.floor(starttime/(T*1e3)))

modifier = np.ones(ym.size)
modifier[startidx:] *= np.linspace(ConstMultStart, ConstMultStop, ym.size-startidx)
#modifier[startidx:] *= np.logspace(np.log10(ConstMultStart), np.log10(ConstMultStop), ym.size-startidx)

ym *= modifier
ym /= np.max(ym)
#ym /= 100

#################################################
#ym[320] = 15

from Onset_Time.PlotOnsetTime import PlotOnsetTime

###
from Onset_Time.Onset_AIC import Onset_AIC

onset, dummy, curves = Onset_AIC(ym, T*1e3)

PlotOnsetTime(ym, curves, T*1e3, onset, xlabel='Time [ms]')

###
from Onset_Time.Onset_Reciprocal_min import Onset_Reciprocal

onset, dummy, curves = Onset_Reciprocal(ym, T*1e3)

PlotOnsetTime(ym, curves, T*1e3, onset, xlabel='Time [ms]')

###
from Onset_Time.Onset_Variance import Onset_Variance

onset, dummy, curves = Onset_Variance(ym, T*1e3)

PlotOnsetTime(ym, curves, T*1e3, onset, xlabel='Time [ms]')
```
