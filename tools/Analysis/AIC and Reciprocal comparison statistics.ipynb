# AIC based and Reciprocal onset time picker comparison statistics

The goal of this document is to compare the properties of the AIC based and Reciprocal onset time picker algorithms.

## Read in preprocessed raw data

```python
from helperfuncs.Raw_Data.ReadAEMeas import ReadAEMeas

dir = '../../../Measurements/BreakTest_AluminiumBeam/results.enhanced/1-0.5'
y_1, y2_1, T, fnames_1, start_time_1 = ReadAEMeas(dir)
dir = '../../../Measurements/BreakTest_AluminiumBeam/results.enhanced/2-0.75'
y_2, y2_2, T, fnames_2, start_time_2 = ReadAEMeas(dir)
dir = '../../../Measurements/BreakTest_AluminiumBeam/results.enhanced/3-0.8'
y_3, y2_3, T, fnames_3, start_time_3 = ReadAEMeas(dir)
dir = '../../../Measurements/BreakTest_AluminiumBeam/results.enhanced/4-1'
y_4, y2_4, T, fnames_4, start_time_4 = ReadAEMeas(dir)
dir = '../../../Measurements/BreakTest_AluminiumBeam2/results2/1-0.1'
y_5, y2_5, T, fnames_5, start_time_5 = ReadAEMeas(dir)
dir = '../../../Measurements/BreakTest_AluminiumBeam2/results2/2-0.15'
y_6, y2_6, T, fnames_6, start_time_6 = ReadAEMeas(dir)
dir = '../../../Measurements/BreakTest_AluminiumBeam2/results2/3-0.2'
y_7, y2_7, T, fnames_7, start_time_7 = ReadAEMeas(dir)
```

```python
y = y_1 + y_2 + y_3 + y_4 + y_5 + y_6 + y_7
y2 = y2_1 + y2_2 + y2_3 + y2_4 + y2_5 + y2_6 + y2_7
fnames = fnames_1 + fnames_2 + fnames_3 + fnames_4 + fnames_5 + fnames_6 + fnames_7
start_time = start_time_1 + start_time_2 + start_time_3 + start_time_4 + start_time_5 + start_time_6 + start_time_7

#y = y_6
#y2 = y2_6
#fnames = fnames_6
#start_time = start_time_6
```

```python
import numpy as np

def trimlength(ys):
    for ii in xrange(len(ys)):
        ys[ii] = ys[ii][:1000]
    return ys
```

```python
from helperfuncs.Raw_Data.removeDC import removeDC

y = trimlength(removeDC(y))
```

## Akaike Information Criterion (AIC) onset time picker

Calculate TDOA using the Akaike Information Criterion (AIC) as onset picker.

```python
from helperfuncs.Onset_Time.CalculateTDOA import CalculateTDOA
from helperfuncs.Onset_Time.Onset_AIC import Onset_AIC

AE_start_AIC, AE_start2_AIC, TD_meas_AIC, qty1_AIC, qty2_AIC = CalculateTDOA(y, y2, T, Onset_AIC)
qty_min_AIC = np.minimum(qty1_AIC, qty2_AIC)
qty_max_AIC = np.maximum(qty1_AIC, qty2_AIC)
```

## Reciprocal method onset time picker

Calculate TDOA using the reciprocal method as onset picker.

```python
from helperfuncs.Onset_Time.CalculateTDOA import CalculateTDOA
from helperfuncs.Onset_Time.Onset_Reciprocal import Onset_Reciprocal

AE_start_rec, AE_start2_rec, TD_meas_rec, qty1_rec, qty2_rec = CalculateTDOA(y, y2, T, Onset_Reciprocal)
qty_min_rec = np.minimum(qty1_rec, qty2_rec)
qty_max_rec = np.maximum(qty1_rec, qty2_rec)
```

## Stats

```python
from helperfuncs.Statistics.Plot_Scatter_Hists import Plot_Scatter_Hists
import numpy as np

Plot_Scatter_Hists( np.concatenate((qty1_AIC, qty2_AIC)), 
                  np.concatenate((qty1_rec, qty2_rec)), 
                  '', 'AIC quality idx []', 'Reciprocal quality idx []', 
                  ylim_min=0, xlim_min=0)
Plot_Scatter_Hists( (np.concatenate((AE_start_AIC, AE_start2_AIC))-np.concatenate((AE_start_rec, AE_start2_rec)))*1e3, 
                  np.concatenate((qty1_rec, qty2_rec)), 
                  '', 'Onset time error [msec]', 'Reciprocal quality idx []', 
                  xlim_min=-0.2, xlim_max=0.2, ylim_min=0)

v_AIC = 3881.32 #m/s - sound speed in aluminium according to AIC
#v_rec = 3723.64 #m/s - sound speed in aluminium according to Reciprocal
v_rec = 4288.08 #m/s - sound speed in aluminium according to Reciprocal
Plot_Scatter_Hists( np.concatenate((AE_start_AIC, AE_start2_AIC))*v_AIC - np.concatenate((AE_start_rec, AE_start2_rec))*v_rec, 
                  np.concatenate((qty1_rec, qty2_rec)), 
                  '', 'Onset time error [m]', 'Reciprocal quality idx []', 
                  xlim_min=-3, xlim_max=3, ylim_min=0)
Plot_Scatter_Hists( np.concatenate((AE_start_AIC, AE_start2_AIC))*v_AIC - np.concatenate((AE_start_rec, AE_start2_rec))*v_rec, 
                  np.concatenate((qty1_AIC, qty2_AIC)), 
                  '', 'Onset time error [m]', 'AIC quality idx []', 
                  xlim_min=-0.5, xlim_max=0.5, ylim_min=0)

Plot_Scatter_Hists( (TD_meas_AIC*v_AIC - TD_meas_rec*v_rec)*1e2, 
                  qty_min_rec, 
                  '', 'distance difference error [cm]', 'Reciprocal quality idx []', 
                  xlim_min=-20, xlim_max=20, ylim_min=0)
Plot_Scatter_Hists( (TD_meas_AIC*v_AIC - TD_meas_rec*v_rec)*1e2, 
                  np.log10(qty_min_rec), 
                  '', 'distance difference error [cm]', 'Reciprocal quality idx []', 
                  xlim_min=-20, xlim_max=20, ylim_min=-6, ylim_max=-0.5)
Plot_Scatter_Hists( (TD_meas_AIC*v_AIC - TD_meas_rec*v_rec)*1e2, 
                  qty_min_AIC, 
                  '', 'distance difference error [cm]', 'AIC quality idx []', 
                  xlim_min=-20, xlim_max=20, ylim_min=0)
Plot_Scatter_Hists( (TD_meas_AIC*v_AIC - TD_meas_rec*v_rec)*1e2, 
                  np.log10(qty_min_AIC), 
                  '', 'distance difference error [cm]', 'AIC quality idx []', 
                  xlim_min=-20, xlim_max=20, ylim_min=-6, ylim_max=-2)
```
