```python
import numpy as np

def removeDC(ys):
    for ii in xrange(len(ys)):
        ys[ii] -= np.mean(ys[ii])
    return ys

def trimlength(ys):
    for ii in xrange(len(ys)):
        ys[ii] = ys[ii][:1000]
    return ys
```

```python
y = trimlength(removeDC(y))
```

## Akaike Information Criterion (AIC) onset time picker

Calculate TDOA using the Akaike Information Criterion (AIC) as onset picker.

```python
from helperfuncs.CalculateTDOA import CalculateTDOA
from helperfuncs.Onset_AIC import Onset_AIC

AE_start_AIC, AE_start2_AIC, TD_meas_AIC, qty1_AIC, qty2_AIC = CalculateTDOA(y, y2, T, Onset_AIC)
qty_min_AIC = np.minimum(qty1_AIC, qty2_AIC)
qty_max_AIC = np.maximum(qty1_AIC, qty2_AIC)
```

## Reciprocal method onset time picker

Calculate TDOA using the reciprocal method as onset picker.

```python
from helperfuncs.CalculateTDOA import CalculateTDOA
from helperfuncs.Onset_Reciprocal import Onset_Reciprocal

AE_start_rec, AE_start2_rec, TD_meas_rec, qty1_rec, qty2_rec = CalculateTDOA(y, y2, T, Onset_Reciprocal)
qty_min_rec = np.minimum(qty1_rec, qty2_rec)
qty_max_rec = np.maximum(qty1_rec, qty2_rec)
```

## Stats

```python
from helperfuncs.PlotScatter2Dhist import PlotScatter2Dhist
import numpy as np

PlotScatter2Dhist( np.concatenate((qty1_AIC, qty2_AIC)), np.concatenate((qty1_rec, qty2_rec)), 'AIC quality idx []', 'Reciprocal quality idx []', ylim_min=0, xlim_min=0)
PlotScatter2Dhist( (np.concatenate((AE_start_AIC, AE_start2_AIC))-np.concatenate((AE_start_rec, AE_start2_rec)))*1e3, np.concatenate((qty1_rec, qty2_rec)), 'Onset time error [msec]', ' Reciprocal quality idx []', xlim_min=-0.2, xlim_max=0.2, ylim_min=0)

v_AIC = 3881.32 #m/s - sound speed in aluminium according to AIC
v_rec = 3723.64 #m/s - sound speed in aluminium according to Reciprocal
PlotScatter2Dhist( np.concatenate((AE_start_AIC, AE_start2_AIC))*v_AIC - np.concatenate((AE_start_rec, AE_start2_rec))*v_rec, np.concatenate((qty1_rec, qty2_rec)), 'Onset time error [m]', ' Reciprocal quality idx []', xlim_min=-3, xlim_max=3, ylim_min=0)
PlotScatter2Dhist( np.concatenate((AE_start_AIC, AE_start2_AIC))*v_AIC - np.concatenate((AE_start_rec, AE_start2_rec))*v_rec, np.concatenate((qty1_AIC, qty2_AIC)), 'Onset time error [m]', ' AIC quality idx []', xlim_min=-0.5, xlim_max=0.5, ylim_min=0)

PlotScatter2Dhist( (TD_meas_AIC*v_AIC - TD_meas_rec*v_rec)*1e2, qty_min_rec, 'distance difference error [cm]', 'Reciprocal quality idx []', xlim_min=-20, xlim_max=20, ylim_min=0)
PlotScatter2Dhist( (TD_meas_AIC*v_AIC - TD_meas_rec*v_rec)*1e2, np.log10(qty_min_rec), 'distance difference error [cm]', 'Reciprocal quality idx []', xlim_min=-20, xlim_max=20, ylim_min=-6, ylim_max=-0.5)
PlotScatter2Dhist( (TD_meas_AIC*v_AIC - TD_meas_rec*v_rec)*1e2, qty_min_AIC, 'distance difference error [cm]', 'AIC quality idx []', xlim_min=-20, xlim_max=20, ylim_min=0)
PlotScatter2Dhist( (TD_meas_AIC*v_AIC - TD_meas_rec*v_rec)*1e2, np.log10(qty_min_AIC), 'distance difference error [cm]', 'AIC quality idx []', xlim_min=-20, xlim_max=20, ylim_min=-6, ylim_max=-2)
```
