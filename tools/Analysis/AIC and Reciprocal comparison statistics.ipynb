```python
y = y_1 + y_2 + y_3 + y_4 + y_5 + y_6 + y_7
y2 = y2_1 + y2_2 + y2_3 + y2_4 + y2_5 + y2_6 + y2_7
fnames = fnames_1 + fnames_2 + fnames_3 + fnames_4 + fnames_5 + fnames_6 + fnames_7
start_time = start_time_1 + start_time_2 + start_time_3 + start_time_4 + start_time_5 + start_time_6 + start_time_7
```

```python
import numpy as np

def removeDC(ys):
    for ii in xrange(len(ys)):
        ys[ii] -= np.mean(ys[ii])
    return ys

def trimlength(ys):
    for ii in xrange(len(ys)):
        ys[ii] = ys[ii][:1000]
    return ys
```

```python
y = trimlength(removeDC(y))
```

## Akaike Information Criterion (AIC) onset time picker

Calculate TDOA using the Akaike Information Criterion (AIC) as onset picker.

```python
from helperfuncs.CalculateTDOA import CalculateTDOA
from helperfuncs.Onset_AIC import Onset_AIC

AE_start_AIC, AE_start2_AIC, TD_meas_AIC, qty1_AIC, qty2_AIC = CalculateTDOA(y, y2, T, Onset_AIC)
qty_min_AIC = np.minimum(qty1_AIC, qty2_AIC)
qty_max_AIC = np.maximum(qty1_AIC, qty2_AIC)
```

## Reciprocal method onset time picker

Calculate TDOA using the reciprocal method as onset picker.

```python
from helperfuncs.CalculateTDOA import CalculateTDOA
from helperfuncs.Onset_Reciprocal import Onset_Reciprocal

AE_start_rec, AE_start2_rec, TD_meas_rec, qty1_rec, qty2_rec = CalculateTDOA(y, y2, T, Onset_Reciprocal)
qty_min_rec = np.minimum(qty1_rec, qty2_rec)
qty_max_rec = np.maximum(qty1_rec, qty2_rec)
```
