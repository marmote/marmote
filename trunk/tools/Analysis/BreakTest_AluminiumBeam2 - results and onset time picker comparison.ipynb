# One measurement

First let's read in the data for BreakTest_AluminiumBeam measurement.

```python
from helperfuncs.ReadAEMeas import ReadAEMeas

#file = '../../../Measurements/BreakTest_AluminiumBeam2/results2/2-0.15/collect_150.bin'
file = '../../../Measurements/BreakTest_AluminiumBeam2/results2/2-0.15/collect_20.bin'
#file = '../../../Measurements/BreakTest_AluminiumBeam/results.enhanced/4-1/collect_106.bin'
y, y2, T, fnames, start_time = ReadAEMeas(file)

Fs = 1/T
```

```python
import numpy as np

def removeDC(ys):
    for ii in xrange(len(ys)):
        ys[ii] -= np.mean(ys[ii])
    return ys

def trimlength(ys):
    for ii in xrange(len(ys)):
        ys[ii] = ys[ii][:600]
    return ys
```

```python
y = trimlength(removeDC(y))
y2 = trimlength(removeDC(y2))
```
