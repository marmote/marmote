# One measurement

First let's read in the data for BreakTest_AluminiumBeam measurement.

```python
from helperfuncs.ReadAEMeas import ReadAEMeas

file = '../../../Measurements/BreakTest_AluminiumBeam/results.enhanced/4-1/collect_106.bin'
y, y2, T, fnames = ReadAEMeas(file)

Fs = 1/T
```

```python
import numpy as np

def removeDC(ys):
    for ii in xrange(len(ys)):
        ys[ii] -= np.mean(ys[ii])
    return ys
```

```python
y = removeDC(y)
y2 = removeDC(y2)

#y[0] = y[0][:200]
#y2[0] = y2[0][:200]
```
