# Anderson–Darling test of normal and uniform pseudo random number generator

## Random normal samples

```python
import numpy as np

N = 1e1
mu = 16
sigma = 13.0

x_norm = np.random.normal(loc=mu, scale=sigma, size=N)
w = np.sqrt(12*sigma**2)
x_uniform = mu + w*np.random.rand(N) - w/2
```

```python
from helperfuncs.Statistics.Plot_Hist import Plot_Hist

alpha = np.array([1.0]) 
mu = np.array([mu]) 
sigma = np.array([sigma])

Plot_Hist(x_norm, 
          'Histogram of normal distribution, N=%d'%(N), 'x' , 'Probability',
          fitcurve=True, alpha=alpha, mu=mu, sigma=sigma)

Plot_Hist(x_uniform, 
          'Histogram of uniform distribution, N=%d'%(N), 'x' , 'Probability',
          fitcurve=True, alpha=alpha, mu=mu, sigma=sigma)
```

## Anderson-Darling test

```python
from scipy.stats import anderson

A2, critical, sig = anderson(x_norm)

print A2
print critical
print sig

A2, critical, sig = anderson(x_uniform)

print A2
print critical
print sig
```
