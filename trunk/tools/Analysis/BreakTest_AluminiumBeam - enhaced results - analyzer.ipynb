### Akaike Information Criterion (AIC)

```python
from helperfuncs.Onset_AIC import Onset_AIC

ym = y[0]
onset, AIC = Onset_AIC(ym, T)

t_y = arange(ym.size)*T
t_AIC = arange(AIC.size)*T

fig, axarr = plt.subplots(2, 1, sharex=True)
axarr = axarr.flat
axarr[0].plot(t_y, ym, 'b', [onset, onset], [np.amin(ym), np.amax(ym)] , 'r')
axarr[1].plot(t_AIC, AIC, 'b', [onset, onset], [np.amin(AIC), np.amax(AIC)], 'r')
```
