# Metropolis Markov chain Monte Carlo sampling of functions

## Introduction

The goal of this notepad is twofold. It examines a method to generate random variables with arbitrary distribution, and secondly, it investigates a way to find maxima (or minima) of arbitrary functions.

For both problems similar approaches are employed. At the core of this the Monte Carlo method is used, which (in a nutshell) randomly picks samples from an examined function to estimate the overall function shape. With the proper sampling this estimate is a good enough description of the original. Samples can be independently generated, or the next sample can always be based upon the previous. In other words, in the latter case the state of the sampling system is always based on the previous state. In that regard the process can be modeled as a Markov chain process. The Metropolis method is an addition to this approach, and it determines when new samples are to be rejected. In the case of arbitrary distribution random variable generation it will reject samples that are less likely to occure given a known Probability Distribution Function (PDF). When finding maxima (or minima), new samples are accepted or rejected based on the function value.   

## Miscellaneous

First the examined function p(x,y) is defined as "p_func".

```python
import numpy as np

lim = [-1, 1] # limits for the function, both x and y

f_func = lambda x, y : 0.5 * np.exp( -90*(x-0.5)**2 - 45*(y+0.1)**4 ) + np.exp( -45*(x+0.4)**2 - 60*(y-0.5)**2 )
p_func = lambda C, x, y : C * np.exp( -f_func(x,y) )
```

```python
def Get_Grid( lim, step_num ):
    step_size = double(lim[1]-lim[0])/step_num
    ls = linspace(lim[0], lim[1], step_num+1)

    X,Y = np.meshgrid(ls[0:-1], ls[0:-1])
    
    return X, Y, step_size, ls 
```

### Normalization

p(x,y) is regarded as a PDF, and as such the underlying area has to be normalized to 1 using the constant C. C is found by evaluating the function in 2D with successively finer grids.

```python
step_num = 1
stop_percent = 0.1

C = 1
C_list = []
grid_points = []

while True:
    step_num = step_num * 2
#    print step_num
    
    X, Y, step_size, _ = Get_Grid( lim, step_num )
#    print X
#    print Y
#    print 'step_size = %f'%(step_size)
    
    grid_points.append(step_num)
    
    Z = p_func(1,X,Y)
#    print Z
    
    prev_C = C
    C = 1/( np.sum(Z) * step_size**2 )
    C_list.append(C)
#    print 'prev_C = %f'%(prev_C)
#    print 'C = %f'%(C)
    
    if np.abs((prev_C - C)/C)*100 <= stop_percent:
        break
        
    if len(C_list) >= 7:
        print 'Error! C not converging fast enough!'
        break
        
print 'Estimate for C: %f\n'%( C )
print 'Grid: %dx%d\n'%( step_num, step_num )
```

## Generating random variables with arbitrary distribution

Given the above PDF first random variables are generated using the Metropolis Markov chain Monte Carlo method.

For an initial guess for the move strategy and step size, first the PDF has to be examined. Apart from the two "holes" at (-0.4, 0.5) and (0.5, -0.1) the PDF can be considered a uniform distribution. This means that there are not too many sharp changes in the function, which would require high resolution sampling. As a consequnce fairly large step sizes can be employed, and the Markov chain process does not have to stay too local. Thus, instead of a move strategy with localized distribution (e.g. Normal) one with a simple uniform distribution was chosen (both along the x and y axes). This has the advantage that uniform random number generators are readily available on most systems. The initial variance was chosen to be 1.

The initial step size of 1 resulted in an acceptance rate of around 32%. This already was a reasonable result according to the literature. But, closer examination revealed that most sample rejections occured due to the samples being out of range (-1,1), not because of the Metropolis MCMC criteria. Hence, the variance (and consequently the step size) was reduced resulting in higher acceptance rates.

```python
from helperfuncs.Statistics.Gen_MMCMC import Gen_MMCMC

step_num = 32
N = 5e5
variance = 1

x, y, accept_rate = Gen_MMCMC( N, variance, p_func_norm, lim_x=lim, lim_y=lim )
print 'Acceptance rate: %.2f%%\n'%(accept_rate)
```

### First moment and the L1 error of PDF and histogram

With the samples generated, simple methods are employed to verify that the samples follow the objective PDF. The first moment (mean) is calculated, and the L1 error of the PDF and histogram are evaluated with increasing sample sizes. Both show convergence.

```python
step_num = 2**5
X, Y, step_size, edges = Get_Grid( lim, step_num )
z_lim = [0, 3]


###################################
# Calculate pdf and moments
pdf = p_func(C,X,Y)
Xl = X[0,:]
Yl = Y[:,0]

#First sum along y, then calculate first moment along x
fmx = np.sum( np.sum(pdf, 0) * step_size**2 * Xl )

#First sum along x, then calculate first moment along y
fmy = np.sum( np.sum(pdf, 1) * step_size**2 * Yl )

print 'Grid: %dx%d\n'%( step_num, step_num )
print 'First moment \"analytical\" estimate along x: %f\n'%( fmx )
print 'First moment \"analytical\" estimate along y: %f\n'%( fmy )



fmx_error = []
fmy_error = []
pdf_error = []
Ns = []
for ii in xrange(1,6):
    ###################################
    # Estimate pdf and moments
    N = 10**ii
    Ns.append(N)
    
    variance = 0.3

    x, y, accept_rate = Gen_MMCMC( N, variance, p_func_norm, lim_x=lim, lim_y=lim )
 
    pdf_estimate, _, _ = np.histogram2d(x, y, bins=(edges, edges), normed=True)
    pdf_estimate = pdf_estimate.transpose()

    fmx_estimate = np.mean(x)
    fmy_estimate = np.mean(y)

#    print 'Grid: %dx%d\n'%( step_num, step_num )
#    print 'First moment estimate along x: %f\n'%( fmx_estimate )
#    print 'First moment estimate along y: %f\n'%( fmy_estimate )


    ###################################
    # Error
    fmx_error.append( (fmx_estimate - fmx) / (lim[1] - lim[0]) * 100 )
    fmy_error.append( (fmy_estimate - fmy) / (lim[1] - lim[0]) * 100 )
    temp = np.abs(pdf_estimate - pdf)
    pdf_error.append( np.sum(temp) )
    
    Plot_2D_PDF_In_3D( X, Y, temp, 
                  title='L1 error, N = %d'%(N), xlabel='x', ylabel='y',
                  z_lim=z_lim)
```

```python
fig, axarr = plt.subplots(1, 1, sharex=True)
axarr.semilogx(Ns, fmx_error,'.-')
axarr.set_title('First moment estimate error along x')
axarr.set_xlabel('Sample number')
axarr.set_ylabel('error [%]')

fig, axarr = plt.subplots(1, 1, sharex=True)
axarr.semilogx(Ns, fmy_error,'.-')
axarr.set_title('First moment estimate error along y')
axarr.set_xlabel('Sample number')
axarr.set_ylabel('error [%]')

fig, axarr = plt.subplots(1, 1, sharex=True)
axarr.semilogx(Ns, pdf_error,'.-')
axarr.set_title('PDF estimate L1 error')
axarr.set_xlabel('Sample number')
axarr.set_ylabel('error')
```

## Finding maxima (or minima) of arbitrary functions

In the following we shall try to find the maximum of f(x,y) employing the normalized PDF as defined above.

In effect the p(x,y) PDF is an "inverse" of the f(x,y) function. This means that p(x,y) gives the highest probability values where f(x,y) has the lowest values. In order to find the maximum of f(x,y) we need the exact opposite: we have to generate samples of f(x,y) densely where the p(x,y) is the lowest. Thus the sample generating MMCMC algorithm has to be slightly modified.

1. Whereas the original MMCMC automatically accepted any new generated sample with a higher probability, the modified algorithm will do the opposite. It will accept generated samples automatically if their probability is lower.

2. Secondly, if the probability of the new sample is not lower, than a uniform random number generator is used to determine whether the new sample gets accepted or not. This is similar to the original MMCMC concept, but the inverse fraction is used, and the uniform random variable is scaled.

```python
#################
# modified MMCMC algorithm
step_num = 64
N = 1e6
variance = 0.3

x, y, accept_rate = Gen_MMCMC( N, variance, p_func_norm,
                              lim_x=lim, lim_y=lim,
                              sample_discard_criterion=lambda x,y : float(y)/x <= 1 and np.random.rand()*3 > float(y)/x )
print 'Acceptance rate: %.2f%%\n'%(accept_rate)
```
