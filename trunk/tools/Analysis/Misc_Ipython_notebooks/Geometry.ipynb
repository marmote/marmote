```python
from sympy import *
```

```python
x_a = Symbol('x_a', real=True, nonnegative=True)
y_a = Symbol('y_a', real=True, nonnegative=True)
z_a = Symbol('z_a', real=True, nonnegative=True)
x_b = Symbol('x_b', real=True, nonnegative=True)
y_b = Symbol('y_b', real=True, nonnegative=True)
z_b = Symbol('z_b', real=True, nonnegative=True)
x_c = Symbol('x_c', real=True, nonnegative=True)
y_c = Symbol('y_c', real=True, nonnegative=True)
z_c = Symbol('z_c', real=True, nonnegative=True)
x_d = Symbol('x_d', real=True, nonnegative=True)
y_d = Symbol('y_d', real=True, nonnegative=True)
z_d = Symbol('z_d', real=True, nonnegative=True)
x_e = Symbol('x_e', real=True, nonnegative=True)
y_e = Symbol('y_e', real=True, nonnegative=True)
z_e = Symbol('z_e', real=True, nonnegative=True)
l = Symbol('l', real=True, nonnegative=True)
d = Symbol('d', real=True, nonnegative=True)
e = Symbol('e', real=True, nonnegative=True)

x_a.assumptions0
```

```python
equations = [Eq(l**2, (x_b-x_c)**2 + (y_b-y_c)**2 + (z_b-z_c)**2), 
             Eq(l**2, (x_b-x_a)**2 + (y_b-y_a)**2 + (z_b-z_a)**2), 
             Eq(l, 8), 
             Eq(x_a, 12.3), 
             Eq(z_a, 1), 
             Eq(y_b, 0), 
             Eq(z_b, 0),
             Eq(x_c, 0),
             Eq(z_c, 1)]
```

```python
equations = [Eq(l, 8),
             Eq(y_a, 4.4),
             Eq(x_a, 12.3), 
             Eq(z_a, 1), 
             Eq(y_b, 0), 
             Eq(z_b, 0),
             Eq(x_c, 0),
             Eq(z_c, 1),
             Eq(l**2, (x_b-x_c)**2 + (y_b-y_c)**2 + (z_b-z_c)**2), 
             Eq(l**2, (x_b-x_a)**2 + (y_b-y_a)**2 + (z_b-z_a)**2), 
             Eq(x_d, x_c + (x_c - x_b)*d/l),
             Eq(y_d, y_c + (y_c - y_b)*d/l),
             Eq(z_d, z_c + (z_c - z_b)*d/l),
             Eq(x_e, x_a + (x_a - x_b)*e/l),
             Eq(y_e, y_a + (y_a - y_b)*e/l),
             Eq(z_e, z_a + (z_a - z_b)*e/l),
             Eq(l**2, (x_e-x_d)**2 + (y_e-y_d)**2 + (z_e-z_d)**2)]
```

```python
solve(equations, [e, x_e, y_e, z_e], manual=True, force=True, minimal=True)
```

```python
solve(equations, [x_a, y_a, z_a, x_b, y_b, z_b, x_c, y_c, z_c, x_d, y_d, z_d, x_e, y_e, z_e, d, e, l], manual=True, force=True)
```

```python
vf, d, a, vi, t = S('vf d a vi t'.split())

equations = [Eq(vf, vi+a*t),
        Eq(d, vi*t + a*t**2/2),
        Eq(a, 10),
        Eq(d, 60),
        Eq(vi, 5)]
```

```python
solve(equations)
```

```python
solve(equations, [a, t, vi, vf, d])
```
